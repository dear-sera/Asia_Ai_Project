## 네이버 뉴스 크롤링

##### 페이지 헤드라인 크롤링해오기


```python
# 모듈 임포트

from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
```

```python
# 카테고리 설정
category = ['Politics', 'Economic', 'Social', 'Culture', 'world', 'IT']
```

```python
# 크롤링할 주소
url = 'https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=100'
```

```python
# 리스폰해오기
headers = {"User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.77 Safari/537.36"}
resp = requests.get(url, headers=headers)
```

```python
# resp list 형식으로 출력
print(list(resp))
```

```python
# resp 타입 확인
print(type(resp))
```

```python
# BeautifulSoup으로 resp을 text로 변환
soup = BeautifulSoup(resp.text, 'html.parser') # html_parser로 파싱
print(soup)
```

```python
# 제목 html 가져오기
title_tags = soup.select('.cluster_text_headline') # class일 경우 앞에 .
print(title_tags)
```

```python
# html에서 제목(헤드라인) 가져오기
titles = []
for title_tag in title_tags:
    # re를 이용하여 comile(세팅)함
    titles.append(re.compile('[^가-힣 | a-z | A-Z | 0-9]+').sub(' ', title_tag.text)) # 한글, 알파벳, 숫자을 제외하고 나머지 자리에는 공백

# 뽑아낸 데이터 및 갯수 확인
print(titles)
print(len(titles))
```

```python
#페이지에 2, 3, 4이렇게 넣어서 이어붙이기

#df_b = pd.DataFrame()
df_page2 = pd.DataFrame(titles, columns=['title'])
df_b = pd.concat([df_b, df_page2], axis='rows', ignore_index=True)
print(df_b.head())
print(df_b.info())
```

```python
# 정치 뉴스 페이지 모두 가져오기
#page_num 순서 = 정치, 경제, 사회, 생활/문화, 세계, IT/과학
page_num = [334, 423, 400, 87, 128, 74]  #각각 분야마다 기사 페이지수를 적어놓기, 사회기사는 730페이지라 너무 많아서 400개로 조절
df_title = pd.DataFrame()

url = 'https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=100#&date=%2000:00:00&page='
for j in range(0, 6):
    df_section_titles = pd.DataFrame()

    for i in range(1, 2):  #1~332까지니까 (1,333)이 맞다. 지금은 다 돌리기 시간없으니깐 시험삼아 숫자변경
      resp = requests.get('https://news.naver.com/main/main.nhn?mode=LSD&mid=shm&sid1=10{0}#&date=%2000:00:00&page={1}'.format(j,i),
                          headers=headers)
      soup = BeautifulSoup(resp.text, 'html.parser')
      title_tags = soup.select('.cluster_text_headline')
      titles = []
      for title_tag in title_tags:
        titles.append(re.compile('[^가-힣 | a-z | A-Z | 0-9 ]').sub(' ', title_tag.text))  #^가 '제외하고'라는 뜻
      df = pd.DataFrame(titles, columns=['title'])
      print('.', end='')  #줄바꾸지말고 . 찍기
      if i % 50 == 0:
        print('\n')  #50개마다 줄바꿈
      df_section_titles = pd.concat([df_section_titles, df],axis='rows', ignore_index=True)
      df_section_titles['category'] = category[j]
    df_title = pd.concat([df_title, df_section_titles], axis='rows', ignore_index=True)
    
print(df_title.head())
print(df_title.info())
```
